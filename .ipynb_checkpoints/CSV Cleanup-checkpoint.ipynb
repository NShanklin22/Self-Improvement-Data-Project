```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("CSV.csv",names=['0','date','programming','gaming','electronics'],index_col=1)
```

```python
df.head()
```

```python
df.drop('0',inplace=True,axis=1)
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df = df.drop(df.index[0],axis=0)
```

```python
df.head()
```

```python
df.head(60)
```

```python
df["programming"] = pd.to_datetime(df['programming'],format="%H:%M:%S")
```

```python
def get_sec(time_str):
    """Get Seconds from time."""
    h, m, s = time_str.split(':')
    return int(h) * 3600 + int(m) * 60 + int(s)
```

```python
for i in range(len(df)):
    if df.iloc[i]['date'] == "0":
         df.iloc[i]['electronics'] = "0:0:0"
    df.iloc[i]['date'] = df.iloc[i]['date']+"/2021"
```

```python
pd.options.mode.chained_assignment = None
df.iloc[47]['programming']
```

```python
df['date'] = pd.to_datetime(df['date'],format='%m/%d/%Y')
```

```python
plt.plot(df.date,df.gaming)
plt.plot(df.date,df.electronics)
plt.plot(df.date,df.programming)
```
