```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import sqlite3
import sqlalchemy
pd.options.mode.chained_assignment = None
```

```python
connection = sqlite3.connect('Data\PersonalData.db')
cursor = connection.cursor()
engine = sqlalchemy.create_engine('sqlite:///Data/PersonalData.db').connect()
```

```python
df = pd.read_csv("Data\PersonalData.csv",index_col=0)
df.head()
```

```python
df = df.drop(df.index[0],axis=1)
df.head(60)
```

```python
df.fillna('0',inplace=True)
```

```python
def get_sec(time_str):
    """Get Seconds from time."""
    h, m, s = time_str.split(':')
    return int(h) * 3600 + int(m) * 60 + int(s)
```

```python
for i in df.index:
    if df.at[i,'programming'] == "0":
         df.at[i,'programming'] = "0:0:0"
    if df.at[i,'gaming'] == "0":
         df.at[i,'gaming'] = "0:0:0"
    if df.at[i,'electronics'] == "0":
         df.at[i,'electronics'] = "0:0:0"
    if df.at[i,'design'] == "0":
         df.at[i,'design'] = "0:0:0"
```

```python
for i in df.index:
     df.at[i,'date'] = str(df.at[i,'date'])+"/2021"
```

```python
df.head()
```

```python
for i in df.index:
    df.at[i,'programming'] = get_sec(df.at[i,'programming'])
    df.at[i,'gaming'] = get_sec(df.at[i,'gaming'])
    df.at[i,'electronics'] = get_sec(df.at[i,'electronics'])
    df.at[i,'design'] = get_sec(df.at[i,'design'])
```

```python
df.head(60)
```

```python
df['date'] = pd.to_datetime(df['date'],format='%m/%d/%Y')
for i in df.index:
    df.at[i,'date'] = datetime.datetime.strftime(df.at[i,'date'],'%m/%d/%y')
print(df['date'])
```
